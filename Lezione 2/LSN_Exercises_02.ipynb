### Exercise 02.1
- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. sampling a uniform distribution in $[0,1]$
2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

<span style="color:red">Show a picture of both your estimations of $I$</span> and their uncertainties with a large number of *throws* $M$ (e.g. $M\ge 10^4$) as a function of the number of blocks, $N$</font>.

## Results

Here we present the results obtained sampling a uniform distriubtion.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
path="Integral/DATA/"
```

```python
M=1000000            # Total number of throws
N=100                 # Number of blocks
L=int(M/N)
x = np.arange(N)
x*=L
```

```python
#SAMPLING A UNIFORM DISTRIBUTION
filename = 'I_uniform.txt'
data = np.loadtxt(path+filename, delimiter='\t', skiprows=1, dtype=float)
#print(data[:,0])
#print(data[:,1])

averages=data[:,0]
errors=data[:,1]

#print(len(averages))
#print(errors)


plt.errorbar(x,averages-1,yerr=errors)
plt.xlabel('#throws')
plt.ylabel(r'$I - I_{true}$')
plt.title("Integral estimation sampling a uniform distribution")
plt.grid(True)
plt.show()
```

The integral approach its expected value within the statistical errors.

Here we show the results obatined with importance sampling. We want to apply importance sampling to reduce the errors of our estimation.

The integrand is $\frac{\pi}{2}\cos(\pi x/2)$. We choose $\frac{n+1}{n}(1-x^n)$ as function from which we sample the random points (with an accept reject method), in the domain \[0,1\]. The n choosen is 2. Basically for increasing n $\frac{n+1}{n}$ goes to 1, and given the domain on which we work $(1-x^n)$ goes to one as well. So a high n will reduce the problem to the uniform sampling case. Instead for $n=2$ we have a function quite similar to the integrand and the normalization $\frac{n+1}{n}$ is maximum at $n=2$, so this should help reduce the error of our estimation.

```python
#Importance sampling
filename = 'I_imp_samp.txt'
data = np.loadtxt(path+filename, delimiter='\t', skiprows=1, dtype=float)
#print(data[:,0])
#print(data[:,1])

averages=data[:,0]
errors=data[:,1]

#print(len(averages))
#print(averages)
#print(errors)

plt.errorbar(x,averages-1,yerr=errors)
plt.xlabel('#throws')
plt.ylabel(r'$I - I_{true}$')
plt.title('Integral estimation with importance sampling')
plt.grid(True)
plt.show()
```

Here is the confrontation between the previous results

```python
#confrontation
filename = 'I_uniform.txt'
uni = np.loadtxt(path+filename, delimiter='\t', skiprows=1, dtype=float)

uni_averages=uni[:,0]
uni_errors=uni[:,1]

filename = 'I_imp_samp.txt'
impsamp = np.loadtxt(path+filename, delimiter='\t', skiprows=1, dtype=float)

impsamp_averages=impsamp[:,0]
impsamp_errors=impsamp[:,1]


plt.errorbar(x,uni_averages-1,yerr=uni_errors, label="uniform sampling")
plt.errorbar(x,impsamp_averages-1,yerr=impsamp_errors, label="importance sampling")
plt.xlabel('#throws')
plt.ylabel(r'$I - I_{true}$')
plt.legend()
plt.title("Confrontation between uniform and importance sampling cases")
plt.grid(True)
plt.show()

#print(impsamp_averages)
#print(uni_errors)
#print(impsamp_errors)
```

Here is a plot to show the reduction of the errors. We plot $\frac{\sigma_{unif}}{\sigma_{imp}}$ versus the number of throws.

```python
appo=[uni_errors[i]/impsamp_errors[i] for i in range(len(uni_errors))]
#print(uni_errors[-1], impsamp_errors[-1], uni_errors[-1]/impsamp_errors[-1])

plt.plot(x, appo)
plt.xlabel('#throws')
plt.ylabel(r'$\frac{\sigma_{unif}}{\sigma_{imp}}$')
plt.title("Reduction of the errors")
plt.grid(True)
plt.show()
```

## Results

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy 
```

```python
path="Rwalk/DATA/"
```

```python
def func(L, x) :
    return L*np.sqrt(x)
```

```python
dim=3

filename="Rw_discrete3.txt"
Rn_disc, err_disc = np.loadtxt(path+filename, delimiter='\t', skiprows=1, unpack=True)

filename="Rw_continuos.txt"
Rn_cont, err_cont = np.loadtxt(path+filename, delimiter='\t', skiprows=1, unpack=True)


Nstep=100;
x=range(0,Nstep)
y=func(1,x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 8))

ax1.plot(x,y, label="theoretical curve")
ax1.errorbar(x, Rn_disc,yerr=err_disc, ecolor="red",label="MC estimation")
ax1.set_xlabel('N step')
ax1.set_ylabel(r"$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$")
ax1.set_title(f"Average distance varying Nstep, discrete case, dimension={dim}")
ax1.grid(True)
ax1.legend()

ax2.plot(x,y, label="tehoretical curve")
ax2.errorbar(x, Rn_cont,yerr=err_cont, ecolor="red", label="MC estimation")
ax2.set_xlabel('N step')
ax2.set_ylabel(r"$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$")
ax2.set_title(f"Average distance varying Nstep, continuous case,  dimension={dim}")
ax2.grid(True)
ax2.legend()

```

As we can see the results obtained both for the discrete and continum case have a power like behaviour of the tipe $a\sqrt(N)$, in our case with $a=1$. This is the kind of behaviour of a diffusive process. It can be seen by solving the diffusion equation for a generic proability distribution function $\rho(x,t)$, where we can imagine x being the end point of a random walk. $\rho(x,t)$ is a gaussian as can be seen by direct solution of the diffusion equation by Fourier transform; this in the case of random walk processes can be understood even by the Central Limit Theorem, since the distnace's distribution is obtained summing over the indipendent random steps. Calculating the square distance average of the distribution $\rho(x,t)$ we find $\langle x^2 \rangle = 2Dt$ which to the case of a random walk process on discrete lattice is equivalent to $a\sqrt(N)$ where $N$ as the function of a discrete time index.

Now we show how this behaviour is universal even regard with the dimension of the lattice. We study it in the discrete case only.
