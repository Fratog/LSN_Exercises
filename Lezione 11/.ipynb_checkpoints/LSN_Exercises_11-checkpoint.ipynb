### Overview 

In this notebook our task will be to perform machine learning regression on noisy data with a Neural Network (NN).

We will explore how the ability to fit depends on the structure of the NN. The goal is also to build intuition about why prediction is difficult.

### The Prediction Problem

Consider a probabilistic process that gives rise to labeled data $(x,y)$. The data is generated by drawing samples from the equation

$$
    y_i= f(x_i) + \eta_i,
$$

where $f(x_i)$ is some fixed, but (possibly unknown) function, and $\eta_i$ is a Gaussian, uncorrelate noise variable such that

$$
\langle \eta_i \rangle=0 \\
\langle \eta_i \eta_j \rangle = \delta_{ij} \sigma
$$

We will refer to the $f(x_i)$ as the **true features** used to generate the data. 

To make predictions, we will consider a NN that depends on its parameters, weights and biases. The functions that the NN can model respresent the **model class** that we are using to try to model the data and make predictions.

To learn the parameters of the NN, we will train our models on a **training data set** and then test the effectiveness of the NN on a *different* dataset, the **validation data set**. The reason we must divide our data into a training and test dataset is that the point of machine learning is to make accurate predictions about new data we have not seen.

To measure our ability to predict, we will learn our parameters by fitting our training dataset and then making predictions on our test data set. One common measure of predictive  performance of our algorithm is to compare the predictions,$\{y_j^\mathrm{pred}\}$, to the true values $\{y_j\}$. A commonly employed measure for this is the sum of the mean square-error (MSE) on the test set:
$$
MSE= \frac{1}{N_\mathrm{test}}\sum_{j=1}^{N_\mathrm{test}} (y_j^\mathrm{pred}-y_j)^2
$$

We will try to get a qualitative picture by examining plots on validation and training data.

### Linear fit

We start by considering the very simple case:
$$
f(x)=2x+1
$$

Let's start defining the parameters of an ideal linear function which we are going to predict through a neural network regression

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

Now let's generate a set of input data which will slightly deviate from our ideal behaviour using a random noise (that actually is set to zero):

```python
import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 500)
x_valid = np.random.uniform(-1, 1, 50)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function (lui dice, cio√® la label)

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
# plot validation and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target, label='target')
plt.scatter(x_valid, y_valid, color='r', label='validation data')
plt.legend()
plt.grid(True)
plt.show()
```

If you remember how a single node of a neural network works, you can easily spot that **just a single neuron can make the job**. So let's start using a simple Sequential model with just one layer on one neuron only!

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
# get a summary of our composed model
model.summary()
```

We are now going to train our model, that is we feed the neuron with the set of training pair x, y_train from which the optimizer will find the best weights to minimize the Mean Square Error loss function (out linear regression function).

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

By looking at weights and biases we can see if the linear fit was successfull: $w_1$ represents the angular coefficient, $b$ the intercept.

```python
# return weights and biases
x, y=model.get_weights()
print(x,y)
print(model.get_weights())
```

```python
# evaluate model
#https://stackoverflow.com/questions/51299836/what-values-are-returned-from-model-evaluate-in-keras
#https://keras.io/api/models/model_training_apis/#evaluate-method
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0]) #val_loss
print('Test accuracy:', score[1]) #val_metrics
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python

# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss']) #mio train loss o Ein
plt.plot(history.history['val_loss']) #mio validation loss o Eout
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
#preparation of the model
my_model = tf.keras.Sequential()
my_model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
#https://keras.io/api/losses/regression_losses/#meansquarederror-class
mse = tf.keras.losses.MeanSquaredError()#this default option should be normalized
my_model.compile(optimizer='sgd', loss=mse)#, metrics=['mse'])
my_model.save_weights('my_model.h5')
```

In order to avoid rebuild our model from scratch for every analysis with different parameters, such as $Nepochs$, we want use the same model. Though at every new fit of the model we must reaload its inital weights  or randomize the intial weights. (https://stackoverflow.com/questions/40496069/reset-weights-in-keras-layer).

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect

def set_variables(N_xtrain, N_xvalid, sig) :
    np.random.seed(1273)
    x_train=np.random.uniform(-1, 1, N_xtrain)
    x_valid=np.random.uniform(-1, 1, N_xvalid)
    print("len(x_train)", len(x_train))
    print("len(x_valid)", len(x_valid))
    x_valid.sort()
    y_target = m * x_valid + b #
    sigma = sig # noise standard deviation, for the moment it is absent
    print("sigma ", sigma)
    y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
    y_valid = np.random.normal(m * x_valid + b, sigma)
    
    return x_train, x_valid, y_train, y_valid

def fit_model(b_size, Nepochs) :
    my_model.load_weights('my_model.h5')
    history = my_model.fit(x=x_train, y=y_train, batch_size=b_size, epochs=Nepochs ,shuffle=True, 
                        validation_data=(x_valid, y_valid))
    return history
```

```python
def plot_param(ax, x, y, xlabel, ylabel ) :
    ax.plot(x, y)#, 
    ax.set_xlabel(f'${xlabel}$')
    ax.set_ylabel(f"${ylabel}$")
    ax.set_title(f"Supervised Learning for linear regression")
    ax.grid(True)
    
def plot_ein_eout(ax, x, y, xlabel, Err ) :
    ax.plot(x, y, marker='o', label=f"{Err}")#, 
    ax.set_xlabel(f'{xlabel}')
    ax.set_ylabel(f"")
    ax.set_title(f"Supervised Learning for linear regression")
    ax.legend()
    ax.grid(True)
```

Now we perform a quick analysis on the in-sample error $E_{in}$ and the out-of-sample error $E_{out}$. In particular we want to see if $E_{out}>E_{in}$ holds. 

```python
sigma=0.5
x_train, x_valid, y_train, y_valid=set_variables(5000, 500, sigma)
#plt.scatter(x_train, y_train)

b_size=32
Nepochs=15
```

```python
my_history=fit_model(b_size, Nepochs) 
```

```python

plt.plot(my_history.history['loss'], label="Train") #mio train loss o Ein
plt.plot(my_history.history['val_loss'], label="Test") #mio validation loss o Eout
plt.title('Supervised Learning for linear regression')
plt.ylabel('Loss')
plt.xlabel('Epoch')
#plt.legend(['Train', 'Test'], loc='best')
plt.xlim(0,10)
#plt.ylim(0, 0.1)
plt.legend()
plt.grid()
plt.show()
```

Generally we expect that the in-sample error $E_{in}$ (i.e. the loss on the training data) to be lower then the out-of-sample error $E_{out}$ (i.e. the loss of the validation data). The reason is natural since the model is fitting its weights, parsing the training data, and using an optimization method to minimize the cost function. So the weights are determined by the loss minimization on the training data. While instead the validation data are unseen. 

In this analysis we have found that $E_{out}>E_{in}$ isn't always satisfied, for this problem. Though there are some points to make for why this happens.

As a first observation we say that in the case $\sigma=0$ (no noise in datasets) we expect this can happen, since there are no fluctuations and we are putting ourselves in a non realistic situation. Adding some noise to get the output for train and validation data, should make the situation more realistic. 

Though, even with $\sigma=0.3$ we still find that $E_{out}>E_{in}$ isn't always satisfied, in particular there is a dependence by seed used to generate the data. Possible reasons could be the following

If Ndata is small I could find $E_{out}<E_{in}$ as a result of the concetration of the validation data around certain points; such a concentration in a particular region can reduce the errors in a random way, since that particular region may be the only part of the domain the algorithm can fit well. If Ndata is large enough to have a uniform sample of the points on the function domain, this problem is avoided. But for Ndata small this could explain the seed dependence. 

Using a larger Ndata of 5000 (training data) and 500 (validation data), and $\sigma=0.5$ we still find some cases where $E_{out}<E_{in}$.  So maybe the fact is that the problem we are dealing with is too easy, in the sense that the low complexity makes trivial the fitting procedure.


After all this stream of consciousness I think the problem is easier respect the way I explained it above. Being the fitting problem an easy one, and having small fluctations the $E_{out}$ and $E_{in}$ converge to the same value, and we can't see an actual point of inversion for the two curves where the validation error becomes greater than the training error. Introducing large fluctations in the datasets the bias of the model becomes relevant and there is a difference between the two type of errors, which sometimes allows to see the point of inversion. 
A more complex fitting problem would present the standard behaviour of the errors we presented in the statistical learning parth of the theory.

```python
def get_data(N_xtrain, N_xvalid, sig) :
    np.random.seed(15)#1273
    x_train=np.random.uniform(-1, 1, N_xtrain)
    x_valid=np.random.uniform(-1, 1, N_xvalid)
    print("len(x_train)", len(x_train))
    print("len(x_valid)", len(x_valid))
    x_valid.sort()
    y_target = pol(x_valid)
    sigma = sig # noise standard deviation, for the moment it is absent
    print("sigma ", sigma)
    y_train = np.random.normal(pol(x_train), sigma) # actual measures from which we want to guess regression parameters
    y_valid = np.random.normal(pol(x_valid), sigma)
    
    return x_train, x_valid, y_target, y_train, y_valid
```

We have checked the predictions of the NN both varying the number of nodes and the complexity of the NN, increasing the number of layers. The following was the model used:

    N_layers=len(N_nodes) #number of layers

    pol_model = tf.keras.Sequential() #sequential

    pol_model.add(Dense(1, input_shape=(1,))) #input which does nothing

    for i in range(N_layers) :
        pol_model.add(Dense(N_nodes[i], activation="relu")) #hidden layers

After some thought there may be some errors in this approach. 

As a first thing, putting the $relu$ activation in the output is probably not a great idea, since will fail to fit a function that has negative values. (I tried to modify the polynomial and everything broke down for this reason).

Even choosing the $relu$ for the hidden layers is probably not the best choiche, since is a discontinuos activation function, and our datasts represent a continuos function. So if the number of nodes is low the prediction won't be smooth, unless we use high numbers of nodes, as we have seen. After these considerations we choose to use a $tanh$ for the hidden layers and no activation function on the output to be surjective in $(-\infty, \infty)$.
So the model which we will use in the following is this:

    pol_model = tf.keras.Sequential() #sequential

    pol_model.add(Dense(1, input_shape=(1,)))

    for i in range(N_layers) :
        pol_model.add(Dense(N_nodes[i], activation="tanh")) #hidden layers


    pol_model.add(Dense(1))#output layers

    pol_model.compile(optimizer='sgd', loss='mse', metrics=['mse'])


## Using different optimizers

Now we want to try to use other optimizers, for this case we used the latter model with hidden layers of 5 and 7 nodes. In the following we define some functions.

```python
def build_model(N_nodes, optim) :
    model = tf.keras.Sequential() #sequential
    for i in range(len(N_nodes)) :
        model.add(Dense(N_nodes[i], activation="tanh")) #hidden layers
   
    model.add(Dense(1))#output layers

    model.compile(optimizer=optim, loss='mse', metrics=['mse'])
    
    return model
```

```python
early=EarlyStopping(monitor='val_loss', min_delta=1e-8, patience=5, verbose=1)

def fit_model(model, b_size, Nepochs, early) :
    history = model.fit(x=x_train, y=y_train, batch_size=b_size, epochs=Nepochs ,shuffle=True, 
                        validation_data=(x_valid, y_valid), callbacks=[early])
    return history

```

```python
def get_history_plot(ax, history, optim) :
    ax.plot(history.history['loss'], label="Train") #mio train loss o Ein
    ax.plot(history.history['val_loss'], label="Test") #mio validation loss o Eout
    ax.set_title(f'Model loss, with optimizer {optim}')
    ax.set_ylabel('Loss')
    ax.set_xlabel('Epoch')
    ax.legend()
    ax.grid()
```

```python
def get_prediction_plot(ax, model, optim) :
    np.random.seed(3)
    x_predicted = np.random.uniform(-1, 1, 100)
    y_predicted = model.predict(x_predicted)
    #ax.scatter(x_train, y_train, color='g', label="training")
    ax.scatter(x_predicted, y_predicted,color='r', label="prediction")
    ax.plot(x_valid, y_target, label="target")
    ax.legend()
    ax.grid(True)
    ax.set_title(f"NN with 2 hidden layers of 5, 7 nodes" + f", optimizer {optim}" )
```

```python
def get_evaluation(model) :
    score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)
    print()
    print('Test loss:', score[0]) #val_loss
    #print('Test accuracy:', score[1]) #val_metrics
```

### Stocastic gradient descent

This is an optimization algorithm based on searching for minima using the gradient of the loss function, a stocastic component is added to ensure more ergodicity in the parameters space. The stocastic component infact helps getting out of local minima, thanks to its randomness. It is the one of the most simple optimization algorithm.

```python
model_sgd=build_model(N_nodes, 'sgd')
history_sgd=fit_model(model_sgd, b_size, Nepochs, early)
```

### Adam optimization algorithm
https://optimization.cbe.cornell.edu/index.php?title=Adam


It is based on the gradient descent idea. It is also an adaptive algorithm, in the sense that varies it's parameters in runtime in order to perform a good search during the whole minimazatoin procedure. For example adapts the length of the move in the parameters space, that is the learning rate of the optimizator. The point is that if we start the optimizatoin with large moves, as we continue to search, the move length should decrease to search on always smaller regions of the parameters space (this is not different from a simulated annealing alogirthm where at temperature varies the step of the metropolis should be adapted to fulfill the 50% empiric rule). This should ensure a faster convergence resepect a plain stocastic gradiaent descent in which there isn't this ability to adapt in runtime. Anyway if the larning rate decrease to fast we risk to lose ergodicity and it can become hard to find a global solution. This faster convergence is evidenced by a comparison of the loss plots for the two optimizers

Adam to adapt the learning rate uses the two first moments of the gradient. With moements we mean the moments of a distribution. The importance of this aspect is the following. The optimization procedure is based on the calculation of the loss function gradient. The gradient is calculated on the single batch of training data, which is a really small portion of the whole dataset, so it is hard that it will carry a complete information always. Thinking of this procedure we can immagine that at every batch analysis we are sampling a value of the gradient from its distribution. As we continue to sample this values of the gradients we have always a more precise knowledge of the destribution and so of it's moments. So Adam uses this values of the moments to choose a direction in the parameter space, and also how to modify the learning rate. 


```python
my_Adam=tf.keras.optimizers.Adam(learning_rate=0.001) 
```

```python

model_Adam=build_model(N_nodes, my_Adam)
history_Adam=fit_model(model_Adam, b_size, Nepochs, early)
```

Try to extend the model to fit a simple trigonometric 2D function such as $f(x,y) = \sin(x^2+y^2)$ in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

```python
import tensorflow as tf
from tensorflow import keras
#import tensorflow.keras.backend as K

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
def fun(x) :
    return np.sin(x[0]**2+x[1]**2)
```

```python
def get_sin_data(N_xtrain, N_xvalid, sig) :
    
    np.random.seed(15)
    xy_train=np.random.uniform(-3/2., 3/2, size=(N_xtrain, 2))
    xy_valid=np.random.uniform(-3/2., 3/2., size=(N_xvalid, 2))
    
    f_target = [ fun(i) for i in xy_train ]
    sigma = sig # noise standard deviation, for the moment it is absent

    f_train=[ np.random.normal(fun(i), sigma) for i in xy_train]
    f_valid=[ np.random.normal(fun(i), sigma) for i in xy_valid]
    
    #i want to use numpy arrays and not pyhton default objects. This should work "like" a copy construct
    return xy_train, xy_valid, np.array(f_target), np.array(f_train), np.array(f_valid) 
```

```python
xy_train, xy_valid, f_target, f_train, f_valid=get_sin_data(5000, 500, 0.25)
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(xy_train[:,0], xy_train[:,1], f_target, label="target")
ax.scatter(xy_valid[:,0], xy_valid[:,1],  f_valid, c="y", label="validation")
plt.xlabel("x")
plt.ylabel("y")
```

```python
N_nodes=[20]
N_layers=len(N_nodes)
b_size=32
Nepochs=100
```

```python
sin_model = tf.keras.Sequential() #sequential

sin_model.add(Dense(10, input_shape=(2,), activation="relu")) #tanh here seemed to perform badly

for i in range(N_layers) :
    sin_model.add(Dense(N_nodes[i], activation="relu")) #hidden layers #tanh here seemed to perform badly

sin_model.add(Dense(1))#output layers


my_Adam=tf.keras.optimizers.Adam()
my_Adagrad=tf.keras.optimizers.Adagrad()
my_Adadelta=tf.keras.optimizers.Adadelta()
sin_model.compile(optimizer=my_Adam, loss='mse', metrics=['mse'])
```

```python

early=EarlyStopping(monitor='val_loss', min_delta=1e-6, patience=5, verbose=1)

history = sin_model.fit(x=xy_train, y=f_train, batch_size=b_size, epochs=Nepochs ,shuffle=True, 
                        validation_data=(xy_valid, f_valid), callbacks=[early])
```

```python
score = sin_model.evaluate(xy_valid, f_valid, batch_size=32, verbose=1)

print()
print('Test loss:', score[0]) #val_loss
print('Test accuracy:', score[1]) #val_metrics
```
