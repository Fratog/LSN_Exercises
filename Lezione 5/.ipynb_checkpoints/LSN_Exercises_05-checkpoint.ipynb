In quantum physics a **wave function**, $\Psi$, is a mathematical description of the state of a quantum system. The wave function is a complex-valued probability amplitude, and the probabilities for the possible results of measurements made on the system can be derived from it.

For now on, let's consider the simple case of a non-relativistic single particle, without spin, in three spatial dimensions. The state of such a particle is completely described by its wave function, $\Psi(\vec{r},t)$, where $\vec{r}$ is position and $t$ is time.

For one spinless particle, if the wave function is interpreted as a probability amplitude, the square modulus of the wave function, $|\Psi(\vec{r},t)|^2$, is interpreted as the probability density that the particle is at $\vec{r}$ at time $t$. Once we have a probability density, we can use Monte Carlo ...

#### Hydrogen atom
The wave functions of the eigenstates of an electron in a Hydrogen atom (this is the only atom for which the Schroedinger equation has been solved exactly) are expressed in terms of spherical harmonics and generalized Laguerre polynomials. It is convenient to use spherical coordinates, and the wave function can be separated into functions of each coordinate:

$$
\Psi_{n,l,m}(r,\theta,\phi)=
\sqrt{\left(\frac{2}{na_0}\right)^3 \frac{(n-l-1)!}{2n[(n+l)!]}}
e^{-r/na_0}\left(\frac{2r}{na_0}\right)^l L_{n-l-1}^{2l+1}\left(\frac{2r}{na_0}\right)
Y_l^m(\theta,\phi)
$$

where $a_0=4\pi\epsilon_0\hbar^2/m_e e^2=0.0529$ nm is the Bohr radius, $L_{n-l-1}^{2l+1}$ are the generalized Laguerre polynomials of degree $n-l-1$, $n=1,2,...$ is the principal quantum number, $l=0,1, ..., n-1$ the azimuthal quantum number, $m=-l, -l+1, ..., l-1, l$ the magnetic quantum number.

For example, the ground state wave function is:
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
whereas one of the three $2p$ excited state is:
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$
<span style="color:blue">Expectation values for the radius turns out to be exactly:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$
</span>

# Equilibration



Before starting to sample the pdf, we must make sure that the Metropolis algorithm is equilibrated. With equilibrated we mean that the algorithm must be in the condition to be ergodic (probably not the best use of the word) right from the start. For example, assume we want to sample the orbital 100 pdf. If we choose an initial point far from the origin then we are in a low probability region of the sample space. So the algorithm will require many Metropolis steps to reach the most probable region of the sample space. This problem can be extremely relevant if we are in low probability regions and the pdf to sample is approximatively flat in that region. 

We now show for the 100 orbital which distances are sampled in the first Metropolis steps, when we choose a initial position far from the origin.



```python
def plot_equil(ax, x, y, string) :
    ax.plot(x, y)
    ax.set_xlabel('step')
    ax.set_ylabel(r"$r $")
    ax.set_title(r"Equilbration, "+ string + " move $\psi_{100}$")
    ax.grid(True)
```

```python
filename="equil_unif_psi100.txt"
unif =np.loadtxt(path + filename ,unpack=True)
xu=np.arange(len(unif))
```

```python
filename="equil_gauss_psi100.txt"
gauss =np.loadtxt(path + filename , unpack=True)
xg=np.arange(len(gauss))
```

```python
fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(14, 10))

plot_equil(ax1, xu, unif, "uniform")

ax2.set_xlim([0,5000])
ax2.set_ylim([0,5]) 
plot_equil(ax2, xu, unif, "uniform")

plot_equil(ax3, xg, gauss, "gaussian")

ax4.set_xlim([0, 5000])
ax4.set_ylim([0,5]) 
plot_equil(ax4, xg, gauss, "gaussian")
```

It is evident how the first seven hundred steps can introduce a bias in our $\langle r \rangle_{100}$ estimate, if we don't peform equilibration. This was even a particularly simple case, as the random walk dindn't have to surpass any barrier to reach the most probable region of the sample space.
To avoid this problem, in our 100 orbital case, is sufficent to start the random walk from the origin, so the algorithm doesn't need equilibration and will sample the pdf correctly from the start. In general we should choose as initial point the maximum of the pdf to sample. The 210 orbital has two peaks, so we have to choose one of them as starting point, and the second one will be sampled when the algorithm is able to surpass the origin. Hence for the 210 orbital case is important to choose a number of Metropolis steps, sufficent to make sure that we are ergodic in the whole sample space.

Another source of possible biases in our estimate is related to the correlation among different sampled points. The Metropolis algorithm sample points according to a Markov process; so indeed there is for sure a correlation among next states of the Markov chain and possibly, a persisting correlation among non consecutive states, that decreases with their distance in the Markov chain. The autocorrelation function of the process helps us understand the characteristic length of memory persistence in our random process. So a study of the autocorrelation function helps us understand after what time scales two states of our stochastic process are not correlated. 

If we obtain an estimate of the average distance from a sample of correlated variables, then our estimate will be bias towards that particular sample. Though, studying the system autocorrelation, we can avoid these problems choosing a block size much larger than the autocorrelation scale. In this way we can reduce the biases in our estimates. The larger the block, the more accurate will be our estimate, but is natural to think that there is a saturation point after which is not really usefull to go. 

We will study the role of autocorellation functions in Metropolis sampling in the next exercisation. For now we use data blocking as always for the values indicated.

# Results

We calculate estimates of the electron's average distance for the hydrogen's atom orbitals 100 and 210. We sample the orbitals squared amplitudes using Metropolis algorithm. We make a confrontation of the results obtained using both uniform and gaussian random walk moves for the transition probability $T(y|x)$ of the Metropolis algorithm.

```python
filename="unif_psi100.txt"
unif_100 , unif_100_err=np.loadtxt(path + filename ,unpack=True)
xu=np.arange(len(unif_100))
```

```python
filename="gauss_psi100.txt"
gauss_100 , gauss_100_err=np.loadtxt(path + filename , unpack=True)
xg=np.arange(len(gauss_100))
```

```python
filename="unif_psi210.txt"
unif_210 , unif_210_err=np.loadtxt(path + filename ,unpack=True)
xu=np.arange(len(unif_210))
```

```python
filename="gauss_psi210.txt"
gauss_210 , gauss_210_err=np.loadtxt(path + filename , unpack=True)
xg=np.arange(len(gauss_210))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

ax1.axhline(y = 1.5 , color ="green", linestyle ="--", label="Theoretical value")
ax1.errorbar(xu, unif_100, yerr=unif_100_err, label="Uniform Tyx")
ax1.errorbar(xg, gauss_100, yerr=gauss_100_err, label="Gaussian Tyx")
ax1.set_xlabel('Blocks')
# ax.set_xscale("log")
ax1.set_ylabel(r"$<r>_{100}$")
ax1.set_title(r"Average distance, $\psi_{100}$ orbital")
ax1.grid(True)
ax1.legend()

ax2.axhline(y = 5 , color ="green", linestyle ="--", label="Theoretical value")
ax2.errorbar(xu, unif_210, yerr=unif_210_err, label="Uniform Tyx")
ax2.errorbar(xg, gauss_210, yerr=gauss_210_err, label="Gaussian Tyx")
ax2.set_xlabel('Blocks')
# ax.set_xscale("log")
ax2.set_ylabel(r"$<r>_{210}$")
ax2.set_title(r"Average distance, $\psi_{210}$ orbital")
ax2.grid(True)
ax2.legend()
```

The Metropolis algorithm is able to predict correct results in both cases.
